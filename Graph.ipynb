```python
import matplotlib.pyplot as plt

# Dados dos modelos
modelos = ['Bi-LSTM', 'Bi-LSTM BN', 'Bi-LSTM BNe LRV', 'Bi-LSTM LRV', 'Bi-LSTM M2']
precisao_bullying = [0.92, 0.91, 0.91, 0.90, 0.90]
precisao_nao_bullying = [0.57, 0.58, 0.57, 0.59, 0.59]
recall_bullying = [0.93, 0.94, 0.95, 0.95, 0.95]
recall_nao_bullying = [0.52, 0.45, 0.43, 0.38, 0.41]
f1_bullying = [0.93, 0.93, 0.92, 0.93, 0.93]
f1_nao_bullying = [0.54, 0.51, 0.49, 0.46, 0.48]
acuracia = [0.87, 0.87, 0.87, 0.87, 0.87]

# Configuração do gráfico
plt.figure(figsize=(10, 6))

# Adicionando linhas para cada métrica
plt.plot(modelos, precisao_bullying, marker='o', label='Precisão (Bullying)')
plt.plot(modelos, precisao_nao_bullying, marker='o', label='Precisão (Não Bullying)')
plt.plot(modelos, recall_bullying, marker='o', label='Recall (Bullying)')
plt.plot(modelos, recall_nao_bullying, marker='o', label='Recall (Não Bullying)')
plt.plot(modelos, acuracia, marker='o', label='Acurácia')

# Adicionando legendas e rótulos
plt.autoscale(axis='x', tight=True)
plt.xlabel('Modelos', fontweight='bold')
plt.ylabel('Métricas')
plt.legend()
plt.grid(True)

# Exibindo o gráfico
plt.show()
```
